```python
def minimum_edit_distance(str1, str2):
    len1 = len(str1)
    len2 = len(str2)

    T = [[0 for j in range(len2 + 1)] for i in range(len1 + 1)]

    for i in range(len1 + 1):
        for j in range(len2 + 1):
            
            if i == 0 and j == 0:
                T[i][j] = 0
                
            elif i == 0:
                T[i][j] = T[i][j - 1] + 1
                
            elif j == 0:
                T[i][j] = T[i - 1][j] + 1
                
            elif str1[i - 1] == str2[j - 1]:
                T[i][j] = T[i - 1][j - 1]
                
            else:
                T[i][j] = 1 + min(
                    T[i][j - 1],
                    T[i - 1][j],
                    T[i - 1][j - 1]
                )

                i, j = len1, len2
    insert = delete = replace = 0

    while i > 0 or j > 0:
        if i > 0 and j > 0 and str1[i - 1] == str2[j - 1]:
            i -= 1
            j -= 1

        elif i > 0 and j > 0 and T[i][j] == T[i - 1][j - 1] + 1:
            replace += 1
            i -= 1
            j -= 1

        elif j > 0 and T[i][j] == T[i][j - 1] + 1:
            insert += 1
            j -= 1

        else:
            delete += 1
            i -= 1

    return T[len1][len2], insert, delete, replace

str1 = input("Enter the String to be convert : ")
str2 = input("Enter the String in which you want to convert : ")

distance, ins, dele, rep = minimum_edit_distance(str1, str2)

print("String 1:", str1)
print("String 2:", str2)
print("Edit Distance:", distance)
print("Insert Count :", ins)
print("Delete Count :", dele)
print("Replace Count:", rep)
```

##### Description :- This code implements minimum edit distance (Levenshtein distance) to support spell checking and correction. It calculates the minimum number of insertions, deletions, and substitutions required to convert one string into another, helping identify the closest correct spelling.
